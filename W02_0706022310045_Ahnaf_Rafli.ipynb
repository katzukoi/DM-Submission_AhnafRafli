# PART A : DATA PROCESSING


```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV from URL source
urld = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv"

# Seperate the coma ","
df = pd.read_csv(urld, sep=",")
```

```python
# Check if df has read the intended URL

df.head()
```

```python
# Ok so it's working now, time to run other details

# Dataframe Info first up
df.info()
```

```python
# Next up, df describe

df.describe()
```

```python
# Looking for null values and formatting them into a dataframe of their own
miss_val = df.isnull().sum()
miss_val_percent = (miss_val/ len(df)) * 100

miss_val_df = pd.DataFrame({ 'Missing Value Count': miss_val, 'Percentage': miss_val_percent})

# Put them into a dataframe for viewing, as well as make the Percentage column view with a % symbol
miss_val_df = miss_val_df[miss_val_df['Missing Value Count'] > 0].sort_values(by='Missing Value Count', ascending=False)
miss_val_df['Percentage'] = miss_val_df['Percentage'].map('{:.2f}%'.format)

# View the dataframe
miss_val_df
```

```python
# Since there's quite a lot of missing values, let's drop some irellevant ones first

# Drop the 'ads_id' and 'property_type' column since it's irrelevant
df.drop('property_type', axis=1, inplace=True)
df.drop('ads_id', axis=1, inplace=True)

# Verify that the column has been dropped
print("Columns after dropping 'property_type':")
print(df.columns)
```

```python
# After that we impute the numerical columns with medians
numerical_cols_median = ['bedrooms', 'bathrooms', 'land_size_m2', 'building_size_m2', 'floors', 'building_age', 'year_built']

for col in numerical_cols_median :
    median_val = df[col].median()
    df[col] = df[col].fillna(median_val)

# Now we go for the categorical columns with mode
categorical_cols_mode = ['certificate', 'property_condition', 'building_orientation', 'furnishing']

for col in categorical_cols_mode :
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)

# Verify that all missing values have been filled
print("Missing values after imputation:")
display(df.isnull().sum())
```

```python
# Ok since the missing values are handled, now we need to see the outliers

# First off we define the numerical columns that are going to be used since these are the most common numericals
# that a outlier will appear in
num_cols_outlier = ['price_in_rp', 'land_size_m2', 'building_size_m2']

# Now we make the Boxplot too visualize our outliers
plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
sns.boxplot(y=df['price_in_rp'])
plt.title('Boxplot of Price (in RP)')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['land_size_m2'])
plt.title('Boxplot of Land Size (m2)')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['building_size_m2'])
plt.title('Boxplot of Building Size (m2)')
```

The reason why i kept the outliers is that they might represent actual legitiamte luxury prices, and removing them might make the data a more skewed and biased that it is

```python
# Lastly we'll go with binning, i'll pick the floors for it
# Bin the 'floors' column into categories

floor_bins = [0, 1, 2, 3, 4, 5, np.inf]
floor_labels = ['1 Floor', '2 Floors', '3 Floors', '4 Floors', '5 Floors']

floor_mapping = {
    1.0: '1 Floor',
    2.0: '2 Floors',
    3.0: '3 Floors',
    4.0: '4 Floors',
    5.0: '5 Floors'
}

df['FloorsCategory'] = df['floors'].map(floor_mapping)

# Display the value counts for the new FloorsCategory column
print("\nValue counts for FloorsCategory:")
display(df['FloorsCategory'].value_counts())
```

# PART B : EDA

```python
# EDA Step 1: Visualize Price Distribution

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['price_in_rp'], kde=True, bins=50) # Using histplot with kde for distribution shape
plt.title('Distribution of House Prices (in RP)')
plt.xlabel('Price (in RP)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```


The histogram shows that the distribution of house prices is highly skewed to the right. This means that the majority of houses are concentrated at lower price ranges, with a long tail extending towards much higher prices. This is typical for real estate data, where there are usually a few very expensive luxury properties that pull the average price up. The presence of this skewness confirms our earlier observation about potential outliers in the [price_in_rp] column. For many statistical models, it might be beneficial to transform the price data (e.g., using a logarithmic transformation) to make it more normally distributed.

```python
# EDA Step 2: Explore Relationship between Price and Building Size

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='building_size_m2', y='price_in_rp', alpha=0.6) # Use scatterplot
plt.title('House Price vs. Building Size')
plt.xlabel('Building Size (m2)')
plt.ylabel('Price (in RP)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

The scatter plot generally shows a positive correlation between building size and price: as the building size increases, the house price tends to increase as well. This is an expected relationship, as larger houses are typically more expensive. However, the relationship is not perfectly linear, and there is a significant spread of prices for any given building size, especially at the lower end of the size scale. This indicates that other factors besides building size also heavily influence the price.

```python
# EDA Step 3: Analyze Price by City

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='city', y='price_in_rp')
plt.title('House Price Distribution by City')
plt.xlabel('City')
plt.ylabel('Price (in RP)')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

The box plot shows that house prices and their distributions vary significantly across the cities. Some cities appear to have generally higher median prices and a wider range of prices (including more high-price outliers) compared to others, indicating potential differences in the real estate market across these locations.

```python
# EDA Step 4: Visualize Distribution of Floors Category

plt.figure(figsize=(8, 6))
# Assign x to hue and set legend=False to fix the FutureWarning
sns.countplot(data=df, x='FloorsCategory', order=df['FloorsCategory'].value_counts().index, palette='viridis', hue='FloorsCategory', legend=False)
plt.title('Distribution of Properties by Number of Floors')
plt.xlabel('Number of Floors')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

The bar chart reveals that most properties in the dataset have either one or two floors, with two-story houses being the most common. Houses with three, four, or five floors are much less frequent. This indicates that one and two-story homes are the dominant types of properties in this dataset, while taller buildings are less common.

```python
# EDA Step 5: Visualize Correlation Matrix with a Heatmap

# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Create a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

The heatmap shows how strongly each pair of numerical features is related. The color and number indicate the strength and direction of the relationship. We can see that house price is most strongly related to building size and land size. Other features like the number of bedrooms and bathrooms also have a positive relationship with price. This suggests that larger properties with more rooms tend to be more expensive.
